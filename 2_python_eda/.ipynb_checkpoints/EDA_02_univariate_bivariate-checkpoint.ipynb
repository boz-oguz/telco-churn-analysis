## Telco Customer Churn
### Univariate & Bivariate Analysis

**Objective**
- Analyze relationships between customer attributes and churn to identify high-risk segments and potential churn drivers.


### 1. Loading libraries and data

```python
# loading libraries

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import missingno as msno

import warnings
warnings.filterwarnings("ignore")
```

```python
# loading data

df = pd.read_csv("../1_dataset/raw_data/WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

### 2. Quick-look

```python
# on data overview we mentioned we need to check TotalCharges, so now we need to fix dtype for quick analysis

df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors = "coerce")
```

```python
# fix SeniorCitizen  since categorical is more useful

df["SeniorCitizen"] = df["SeniorCitizen"].map({0: "No", 1: "Yes"})
df.head()
```

```python
numeric_cols = ["tenure", "MonthlyCharges", "TotalCharges"]
df[numeric_cols].describe()
```

### 3. Univariate & Bivariate Analysis

```python
# lets understand Churn distribution

churns = df["Churn"].value_counts(normalize = True) * 100

churns.plot(kind = "bar")
plt.title(" Churn Distribution %")
plt.ylabel("Percentage %")
plt.show()
```

- Churn affects a significant minority of customers.
- Even small reductions in churn would have meaningful revenue impact.

```python
# lets see impact of Contract Type on Churn

contract_churn = (
    df.groupby("Contract")["Churn"]
    .value_counts(normalize = True)
    .rename("churn_rate")
    .reset_index()
)

contract_churn = contract_churn[contract_churn["Churn"] == "Yes"]

sns.barplot(data = contract_churn, x = "Contract", y = "churn_rate", palette = "Set2")
plt.title("Churn Rate by Contract Type")
plt.ylabel("Churn Rate")
plt.show()
```

- Month-to-month contracts show substantially higher churn.
- Longer contracts act as a natural retention mechanism.

```python
# Churn by tenure

plt.figure(figsize = (8,6))
sns.histplot(data = df, x = "tenure", hue = "Churn", bins = 30, multiple = "stack", palette = "Set2")
plt.title("We are losing customers on below tenure...")
plt.show()
```

- Churn is concentrated among customers with low tenure.
- Early lifecycle engagement appears critical.

```python
# now observe monetary impact on Churn

plt.figure(figsize = (8,6))
sns.boxplot(data = df, x = "Churn", y = "MonthlyCharges", palette = "Set2")
plt.title("Monthly Charge range that we lose customers on...")
plt.show()
```

- Customers who churn tend to have higher monthly charges.
- Price sensitivity may be a contributing factor.

```python
# understand if senority impact equation on Churn

senior_churn = (
    df.groupby("SeniorCitizen")["Churn"]
    .value_counts(normalize = True)
    .rename("rate")
    .reset_index()
)

senior_churn =  senior_churn[senior_churn["Churn"] == "Yes"]

sns.barplot(data = senior_churn, x = "SeniorCitizen", y = "rate", palette = "Set2")
plt.title("How age shows impact on churn...")
plt.show()
```

- Senior customers show different churn behavior.
- Targeted communication strategies may be required.
