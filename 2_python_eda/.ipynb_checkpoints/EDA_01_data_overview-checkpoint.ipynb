##### The data set includes information about:

### Initial Observation

### 4. Missing & Invalid Values

```python
df.isnull().sum()
```

```python
# it seems like there is no null value in our dataset, however TotalCharges dtype is object even though it has numeric values

df[df["TotalCharges"] == " "].shape
```

```python
# lets see if we can find those values

df[df["TotalCharges"] == " "][["tenure", "MonthlyCharges", "TotalCharges"]]
```

### Data Quality Note
- `TotalCharges` contains blank values rather than nulls.
- These cases correspond to customers with very low tenure.
- This will require cleaning before SQL loading and modeling.
